# 1. Create a function in python to read the text file and replace specific content of the file.

```python
#Replace function

def replace(old_value,new_value):
    
    f = open("example.txt", "r")
    content = f.read()
    #l1 list is for old content and l2 list is for new replace content
    l1 = []
    l2 = []
    
    #appending old sentence into l1
    l1.append(content) 
    
    #split l1 first value
    string_Split = l1[0].split() 
    
    #for loop to access old value from l1
    for i in string_Split:
        
        #if-else condition to replace old_value to new_value
        if i == old_value:
            l2.append(new_value)
        else:
            l2.append(i)
    #convert list to string
    list_to_str = " ".join(map(str,l2))
    return list_to_str
replace("placement","screnning")
```

```python
#Replace content using tkinter

#open a text file 
from tkinter import *
from tkinter import ttk
import tkinter as tk

# main window
base1 =Tk()
base1.geometry("420x360+380+85")
base1.title("Replace_word")

# "Enter file name" label Widget
Label1 = Label(base1 , text = "Enter file name" , font = ("Georgia" , "13"),fg= "#3A3B3C")
Label1.place(x = 20 , y = 17)

# File name Entry Widget
Ent1 = Entry(base1, width = 20,bg= "white",font = ("Arial" , "10"))
Ent1.place(x = 175 , y = 19)

#Entry for old content Widget
Ent2 = Entry(base1, width = 18,bg= "white",font = ("Arial" , "11"))
Ent2.place(x = 8 , y = 60)

#"replace by" label Widget
Label1 = Label(base1 , text = "replace by" , font = ("Georgia" , "13"),fg= "#3A3B3C")
Label1.place(x = 165 , y = 58)

#Entry Widget for new content
Ent3 = Entry(base1, width = 18,bg= "white",font = ("Arial" , "11"))
Ent3.place(x = 260 , y = 60)

#Text Widget to show file data
T = Text(base1, height = 10, width = 50)
T.place(x=10 , y = 105)

#Show file function
def show_file():
    #Get file name from Entry
    file_name = str(Ent1.get())
    
    #open given file
    open_file = open(file_name, "r")
    
    #read given file
    content = open_file.read()
    
    #insert file data into text
    T.insert(tk.END, content)

#Replace function
def replace():
    
    #get file name
    file_name = str(Ent1.get())
    
    #open given file into read mode
    f = open(file_name, "r")
    
    #read given file
    content = f.read()
    
    #get old and new value
    old_value = str(Ent2.get()) 
    new_value = str(Ent3.get())
    
    #l1 list is for old content and l2 list is for new replace content
    l1 = []
    l2 = []
    
    #appending old sentence into l1
    l1.append(content) 
    
    #split l1 first value
    string_Split = l1[0].split() 
    
    #for loop to access old value from l1
    for i in string_Split:
        
        #if-else condition to replace old_value to new_value
        if i == old_value:
            l2.append(new_value)
        else:
            l2.append(i)
    #convert list to string
    list_to_str = " ".join(map(str,l2))
    
    #open given file into write mode
    f = open(file_name, "w")
    f.write(list_to_str)
    f.close()
    
    #open and read the file after the appending
    f = open(file_name, "r")
    print(f.read())
    
    #read new data file
    new_content = f.read()
    
    #delete old data
    T.delete("1.0","end")
    
    #insert new data into text
    T.insert(tk.END, new_content)
    

#Button for open_file
b2 = Button(base1, text = "open_file" ,width = "14", background = "Green" , foreground = "White",font = ("Arial Bold",10),command = show_file)
b2.place(x = 70 , y = 297)

#Button for Replace_content
b3 = Button(base1, text = "Replace_content" ,width = "14", background = "orange" , foreground = "Black",font = ("Arial Bold",10),command = replace)
b3.place(x = 240 , y = 297)

#Display the window without pressing key
base1.mainloop()
```
